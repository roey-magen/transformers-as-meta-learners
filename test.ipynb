```python
import os
import glob
from PIL import Image
import matplotlib.pyplot as plt

# Directories
input_dir = 'plots'
output_dir = 'latest_plots'

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Step 1: Get all PNG files from the 'plots' directory
image_files = glob.glob(os.path.join(input_dir, '*.png'))

# Step 2: Sort by modification time (latest first)
image_files.sort(key=os.path.getmtime, reverse=True)

# Step 3: Take the 3 most recently modified images
latest_images = image_files[:3]

# Step 4: Display them in a single figure
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, img_path in enumerate(latest_images):
    img = Image.open(img_path)
    axs[i].imshow(img)
    axs[i].axis('off')
    

plt.tight_layout()

# Step 5: Save the figure
output_path = os.path.join(output_dir, 'latest_combined_plot.png')
plt.savefig(output_path)
plt.show()

print(f"Saved combined plot to: {output_path}")
```
